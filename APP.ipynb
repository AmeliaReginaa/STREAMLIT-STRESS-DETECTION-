```python
import streamlit as st
import numpy as np
from PIL import Image
from tensorflow.keras.models import load_model

# ===== LOAD MODEL =====
model = load_model("/content/stress_detection_model.h5")

# ===== PREPROCESS =====
def preprocess(image):
    image = image.resize((48,48))  # ubah kalau model beda
    image = image.convert('L')     # grayscale
    image = np.array(image) / 255.0
    image = np.reshape(image, (1,48,48,1))
    return image

# ===== UI =====
st.title("🧠 Stress Detection App")
st.write("Upload gambar wajah untuk mendeteksi stres")

uploaded_file = st.file_uploader("Upload Gambar", type=["jpg","png","jpeg"])

if uploaded_file:
    image = Image.open(uploaded_file)
    st.image(image, caption="Gambar", use_column_width=True)

    img = preprocess(image)
    prediction = model.predict(img)

    confidence = float(np.max(prediction)) * 100
    label = np.argmax(prediction)

    st.subheader("Hasil Prediksi")

    if label == 1:
        st.error(f"STRESS 😟 ({confidence:.2f}%)")
    else:
        st.success(f"TIDAK STRESS 😊 ({confidence:.2f}%)")

    st.progress(int(confidence))
```
